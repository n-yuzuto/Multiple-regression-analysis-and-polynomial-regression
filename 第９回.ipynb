```python
#課題1.1
from sklearn import datasets
X,y = datasets.load_boston(return_X_y=True)
x = X[:,(5,10)]
```

```python
from sklearn.linear_model import LinearRegression
clf = LinearRegression()
clf.fit(x,y)
print(clf.intercept_, clf.coef_) #theta_0. theta_1
```

```python
ypred = clf.predict(x)
ymean = np.mean(y)
n = x.shape[0]
d = x.shape[1]
nu_e = n-d-1
S_Residual = np.sum((y-ypred)**2)
S_Total = np.sum((y-ymean)**2)
R_Squared = 1-S_Residual / S_Total
R_Adjusted = 1 - ((S_Residual/nu_e) / (S_Total / (n-1)))
print("決定係数              :", R_Squared)
print("自由度調整済み決定係数:", R_Adjusted)
```

```python
plt.figure(figsize=(4,4))
plt.scatter(ypred,y)
plt.axis("equal")
plt.grid()
plt.xlabel("$\hat{y}$")
plt.ylabel("y")
```

```python
#課題2.3
from sklearn import datasets
from sklearn.linear_model import LinearRegression
import numpy as np
import matplotlib.pyplot as plt
#散布図
X,y = datasets.load_boston(return_X_y=True)
x = X[:,(12,)]
plt.scatter(x,y)
#一次式の直線
clf = LinearRegression()
clf.fit(x,y)
print(clf.intercept_, clf.coef_) #theta_0. theta_1
xmin = np.min(x)
xmax = np.max(x)
xarray = np.linspace(xmin, xmax, 100).reshape((-1,1))
yarray = clf.predict(xarray)
plt.plot(xarray, yarray, color="orange")
#２次式
X,y = datasets.load_boston(return_X_y=True)
x1 = np.hstack((X[:, (12,)], X[:, (12,)]**2))
clf = LinearRegression()
clf.fit(x1,y)
print(clf.intercept_, clf.coef_) #theta_0. theta_1 theta_2
xarray1 = np.hstack((xarray,xarray**2))
yarray1 = clf.predict(xarray1)
plt.plot(xarray,yarray1, color = "red")
plt.xlabel("x")
plt.ylabel("y")
plt.xlim(xmin, xmax)
```
